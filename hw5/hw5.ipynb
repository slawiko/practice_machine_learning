### Задание №1

Параметр k влияет на то, на сколько частей будет разбита выборка, чтобы потом каждую из этих частей использовать как валидационную выборку. 
Параметр k обратно пропорционален размеру валидационной выборки. Если количество элементов в выборке велико, параметр k можно выбирать поменьше, потому что в таком случае размер тренировочной выборки не будет страдать (как он страдал бы в случае маленького k для маленькой выборки).

### Задание №2

На тренировочной выборке обучаются, на валидационной выборке корректируют модель, на тестовой проверяют true-risk. Скорее всего в практических задачах не желают жертвовать ни проверкой true-risk ни корректировкой модели. 

### Задание №3

Недостатки:
- затраты времени. Перебрать все возможные разбиения выборки - O(2^n)

Преимущества:
- возможность выбора наилучшего соотношения тестовой и тренировочной выборки
- более усреднённая ошибка, а значит более точная оценка качества
